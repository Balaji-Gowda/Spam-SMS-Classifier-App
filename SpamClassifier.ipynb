```python
from sklearn.metrics import accuracy_score,classification_report
# from wordcloud import WordCloud
```

```python
from nltk.tokenize import word_tokenize
```

```python
from nltk.corpus import stopwords
```

```python
stp=stopwords.words('english')
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
from nltk.stem import SnowballStemmer
ss=SnowballStemmer("english")
```

```python
df=pd.read_csv('Datasets/spamSMSKaggle.csv',usecols=['v1','v2'])
df.head()
```

```python
df.columns=['LABEL','MESSAGE']
```

```python
df.head()
```

```python
df.shape
```

```python
df['length']=df['MESSAGE'].apply(len)
```

```python
def cleaning(strr):
    txt=re.sub('[^a-zA-Z]',' ',strr)
    txt=txt.lower()
    txt=txt.split()
    words=[ss.stem(word) for word in txt if word not in stp]
    return " ".join(words)
```

```python
df['MESSAGE']=df['MESSAGE'].apply(cleaning)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
x=cv.fit(df['MESSAGE'])
```

```python
x.vocabulary_
```

```python
x.get_feature_names()
```

```python
x=cv.fit_transform(df['MESSAGE']).toarray()
```

```python
x
```

```python
df_array=pd.DataFrame(x,columns=cv.get_feature_names())
df_array.head()
```

```python
df['LABEL']=df['LABEL'].map({'ham':1,'spam':0})
```

```python
y=pd.DataFrame(df['LABEL'])
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_array,y,test_size=0.2,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mod=MultinomialNB().fit(x_train,y_train)
```

```python
y_pred=mod.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
y_pred
```

```python
print(classification_report(y_test,y_pred))
```
